**Compute the Number of Times a Pattern Appears in a Text
Rosalind Problem 10**

```python
# Rosalind Problem: BA1A - Count Pattern Occurrences
# Reads from "rosalind_ba1a.txt" and prints the count.

# Step 1: Upload the file in Google Colab
from google.colab import files
uploaded = files.upload()  # Upload your file named "rosalind_ba1a.txt"

# Step 2: Read input from the file
with open("rosalind_ba1a.txt", "r") as f:
    lines = f.read().strip().split()
    Text = lines[0]     # First line is the DNA Text
    Pattern = lines[1]  # Second line is the Pattern

# Step 3: Initialize the count
count = 0
n = len(Text)
m = len(Pattern)

# Step 4: Slide a window across Text to find matches
for i in range(n - m + 1):
    if Text[i:i+m] == Pattern:
        count += 1

# Step 5: Print the result
print(count)
```

**Rosalind Problem 11**

```python
from collections import defaultdict
import io
import ipywidgets as widgets
from IPython.display import display

def frequent_words(text, k):
    kmer_count = defaultdict(int)
    for i in range(len(text) - k + 1):
        kmer = text[i:i+k]
        kmer_count[kmer] += 1
    max_count = max(kmer_count.values())
    return [kmer for kmer, count in kmer_count.items() if count == max_count]

# File upload widget
upload = widgets.FileUpload(
    accept='.txt',  # Accept text files
    multiple=False  # Only one file at a time
)
display(upload)

def process_upload(change):
    if upload.value:
        # Get the uploaded file content
        uploaded_file = list(upload.value.values())[0]
        content = io.StringIO(uploaded_file['content'].decode('utf-8'))
        lines = content.read().splitlines()
        text = lines[0].strip()  # First line: DNA string
        k = int(lines[1].strip())  # Second line: integer k
        result = frequent_words(text, k)
        print("Most frequent k-mers:", " ".join(result))

# Observe changes when a file is uploaded
upload.observe(process_upload, names='value')
```

Rosalind Problem 12 Clump Finding Problem

```python
from collections import defaultdict
import io
import ipywidgets as widgets
from IPython.display import display

def clump_finding(genome, k, L, t):
    clumps = set()
    n = len(genome)

    # Frequency dictionary for k-mers in the first window
    freq_map = defaultdict(int)

    # First window
    for i in range(L - k + 1):
        kmer = genome[i:i+k]
        freq_map[kmer] += 1
    for kmer, count in freq_map.items():
        if count >= t:
            clumps.add(kmer)

    # Slide window
    for i in range(1, n - L + 1):
        # Remove the first k-mer of the previous window
        first_kmer = genome[i-1:i-1+k]
        freq_map[first_kmer] -= 1

        # Add the new k-mer of the current window
        new_kmer = genome[i+L-k:i+L]
        freq_map[new_kmer] += 1

        if freq_map[new_kmer] >= t:
            clumps.add(new_kmer)

    return clumps


# File upload widget
upload = widgets.FileUpload(
    accept='.txt',  # Accept text files
    multiple=False
)
display(upload)

def process_upload(change):
    if upload.value:
        # Get uploaded file content
        uploaded_file = list(upload.value.values())[0]
        content = io.StringIO(uploaded_file['content'].decode('utf-8'))
        lines = content.read().splitlines()

        genome = lines[0].strip()        # First line = Genome string
        k, L, t = map(int, lines[1].split())  # Second line = k L t

        result = clump_finding(genome, k, L, t)
        print("Clumps found:", " ".join(result))

# Run when file is uploaded
upload.observe(process_upload, names='value')
```

Rosalind 13 Minimum Skew Problem

```python
import io
import ipywidgets as widgets
from IPython.display import display

def minimum_skew(genome):
    skew = [0]  # Skew at position 0
    for base in genome:
        if base == "G":
            skew.append(skew[-1] + 1)
        elif base == "C":
            skew.append(skew[-1] - 1)
        else:
            skew.append(skew[-1])

    min_val = min(skew)
    positions = [i for i, val in enumerate(skew) if val == min_val]
    return positions


# File upload widget
upload = widgets.FileUpload(
    accept='.txt',  # Only allow text files
    multiple=False
)
display(upload)

def process_upload(change):
    if upload.value:
        # Read uploaded file
        uploaded_file = list(upload.value.values())[0]
        content = io.StringIO(uploaded_file['content'].decode('utf-8'))
        genome = content.read().strip()

        result = minimum_skew(genome)
        print("Positions with minimum skew:", " ".join(map(str, result)))

# Trigger when file is uploaded
upload.observe(process_upload, names='value')
```

Rosalind 15 Generate the d-Neighborhood of a String

```python
import io
import ipywidgets as widgets
from IPython.display import display

def hamming_distance(p, q):
    """Compute Hamming distance between two equal-length strings."""
    return sum(1 for a, b in zip(p, q) if a != b)

def neighbors(pattern, d):
    """Generate the d-neighborhood of a DNA pattern."""
    if d == 0:
        return {pattern}
    if len(pattern) == 1:
        return {"A", "C", "G", "T"}

    neighborhood = set()
    suffix_neighbors = neighbors(pattern[1:], d)
    for text in suffix_neighbors:
        if hamming_distance(pattern[1:], text) < d:
            for x in "ACGT":
                neighborhood.add(x + text)
        else:
            neighborhood.add(pattern[0] + text)
    return neighborhood


# File upload widget
upload = widgets.FileUpload(
    accept='.txt',  # Only text files
    multiple=False
)
display(upload)

def process_upload(change):
    if upload.value:
        # Get uploaded file content
        uploaded_file = list(upload.value.values())[0]
        content = io.StringIO(uploaded_file['content'].decode('utf-8'))
        lines = content.read().splitlines()

        pattern = lines[0].strip()
        d = int(lines[1].strip())

        result = neighbors(pattern, d)
        print("\n".join(sorted(result)))  # Print neighbors, one per line

# Trigger when file is uploaded
upload.observe(process_upload, names='value')
```

Rosalind 16 Approximate Pattern Matching Problem

```python
import io
import ipywidgets as widgets
from IPython.display import display

def hamming_distance(p, q):
    """Compute Hamming distance between two equal-length strings."""
    return sum(1 for a, b in zip(p, q) if a != b)

def approximate_pattern_matching(pattern, text, d):
    """Find all starting positions where pattern occurs in text with  d mismatches."""
    positions = []
    k = len(pattern)
    for i in range(len(text) - k + 1):
        substring = text[i:i+k]
        if hamming_distance(pattern, substring) <= d:
            positions.append(i)
    return positions


# File upload widget
upload = widgets.FileUpload(
    accept='.txt',  # Only text files
    multiple=False
)
display(upload)

def process_upload(change):
    if upload.value:
        # Get uploaded file content
        uploaded_file = list(upload.value.values())[0]
        content = io.StringIO(uploaded_file['content'].decode('utf-8'))
        lines = content.read().splitlines()

        pattern = lines[0].strip()
        text = lines[1].strip()
        d = int(lines[2].strip())

        result = approximate_pattern_matching(pattern, text, d)
        print(" ".join(map(str, result)))

# Trigger when file is uploaded
upload.observe(process_upload, names='value')
```

**Rosalind 17 Frequent Words with Mismatches Problem**

```python
from google.colab import files

def hamming_distance(p, q):
    return sum([p[i] != q[i] for i in range(len(p))])

def neighbors(pattern, d):
    if d == 0:
        return {pattern}
    if len(pattern) == 1:
        return {"A", "C", "G", "T"}
    neighborhood = set()
    suffix_neighbors = neighbors(pattern[1:], d)
    for text in suffix_neighbors:
        if hamming_distance(pattern[1:], text) < d:
            for nucleotide in "ACGT":
                neighborhood.add(nucleotide + text)
        else:
            neighborhood.add(pattern[0] + text)
    return neighborhood

def frequent_words_with_mismatches(text, k, d):
    freq_map = {}
    n = len(text)
    for i in range(n - k + 1):
        pattern = text[i:i+k]
        neighborhood = neighbors(pattern, d)
        for approx in neighborhood:
            freq_map[approx] = freq_map.get(approx, 0) + 1
    max_count = max(freq_map.values())
    return [kmer for kmer, count in freq_map.items() if count == max_count]


# --- Upload and run in Colab ---
uploaded = files.upload()   # lets you upload your file

file_name = list(uploaded.keys())[0]  # get the uploaded file name

with open(file_name, "r") as f:
    text = f.read().strip()

print("DNA sequence loaded from:", file_name)

# Ask user for k and d
k = int(input("Enter k (k-mer length): "))
d = int(input("Enter d (maximum mismatches): "))

result = frequent_words_with_mismatches(text, k, d)
print("Most frequent k-mers with up to", d, "mismatches:")
print(" ".join(result))
```
